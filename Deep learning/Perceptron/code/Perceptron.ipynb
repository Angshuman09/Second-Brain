```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('assets/passExam.csv')
```

```python
print(df.shape)
df.head()
```

```python
sb.scatterplot(x=df['sleep_hours'], y=df['coffee_cups'], hue=df['passed_exam'])
```

```python
X = df.iloc[:,0:2]
Y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,Y)
```

```python
p.coef_
```

```python
p.intercept_ #w1x1 + w2x2 + b
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, Y.values, clf=p, legend=2)
```
